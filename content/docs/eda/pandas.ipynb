## Reshape

```python
mw = oo[(oo.Edition == 2008) & ((oo.Event == '100m') | (oo.Event == '200m'))]
mw.head()
```

```python
g = mw.groupby(['NOC', 'Gender', 'Discipline', 'Event']).size()
```

```python
df = g.unstack(['Discipline', 'Event'])
df
```

```python
df.stack()
```

```python
df.unstack()
```

#### Plot the number of gold medals won by US male and female athletes throughout Olympic history

```python
gm = oo[(oo.Medal=='Gold') & (oo.NOC == 'USA')]
gm.groupby(['Edition', 'Gender']).size().unstack('Gender', fill_value=0).plot(kind='bar')
```

```python
gm.groupby(['Edition', 'Gender']).size().unstack('Gender', fill_value=0).plot()
```

#### Plot 5 athletes with most gold medals in Olympic history, considering silver and bronze for ties

```python
g = oo.groupby(['Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
```

```python
g.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']].head(5).plot(kind='bar')
```
